```python
from gensim.models import KeyedVectors
import pickle
import numpy as np
```

```python
word_vectors = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/Minor Project 2/GoogleNews-vectors-negative300.bin',binary=True)
```

```python
print(word_vectors['the'].shape)
print(word_vectors['the'])
```

```python
with open("/content/drive/MyDrive/Minor Project 2/data files4/word_to_idx.pkl","rb") as w2i:
    word_to_idx=pickle.load(w2i)
    
# index 0 is for cliping as if dont have words then max words we will place 0 there
vocab_size = len(word_to_idx) + 1
print("vocab size",vocab_size)
```

```python
def get_embedding_matrix():
    skip=0
    emb_dim = 300
    matrix = np.zeros((vocab_size,emb_dim))

    for word,idx in word_to_idx.items():
        
        if word in word_vectors.vocab:
            matrix[idx] = np.array(word_vectors[word],dtype='float')
        else:
            skip+=1
            print(word)
    print(skip)
    return matrix
```

```python
embedding_matrix = get_embedding_matrix()
embedding_matrix.shape
```
